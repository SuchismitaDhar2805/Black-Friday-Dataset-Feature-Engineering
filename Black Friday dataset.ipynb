# Feature Engineering

```python
Final_dataset=Final_dataset.drop(columns= ['User_ID'])
```

### Feature Selection: According to the domain knowledge the user id is the less important feature for train the model

#### Data Preprocessing of the different catagorical features

```python
Final_dataset['Gender']=Final_dataset['Gender'].map({'M':0,'F':1})
```

```python
Final_dataset
```

```python
Final_dataset['Age'].unique()
```

```python
Final_dataset['Age']=Final_dataset['Age'].map({'0-17':1,'18-25':2,'26-35':3,'36-45':4,'46-50':5,'51-55':6,'55+':7})
```

```python
Final_dataset.head()
```

```python
Final_dataset['City_Category'].unique()
```

```python
Final_dataset['City_Category']=Final_dataset['City_Category'].map({'A':1,'B':2,'C':3})
```

```python
Final_dataset.head()
```

```python
Final_dataset['Stay_In_Current_City_Years'].unique()
```

```python
Final_dataset['Stay_In_Current_City_Years']=Final_dataset['Stay_In_Current_City_Years'].str.replace('+',' ')
```

```python
Final_dataset.head()
```

```python
Final_dataset.dtypes
```

```python
Final_dataset['Stay_In_Current_City_Years']=Final_dataset['Stay_In_Current_City_Years'].astype(int)
```

```python
Final_dataset.dtypes
```

#### Handling the missing values

```python
Final_dataset.isnull().sum()
```

```python
Final_dataset['Product_Category_2'].unique()
```

#### In the Product_Category_2, the values are the descrete type. So the best option to fill the missing values with the mode of this column. 

```python
Final_dataset['Product_Category_2'].value_counts()
```

```python
Final_dataset['Product_Category_2'].mode()
```

```python
Final_dataset['Product_Category_2']=Final_dataset['Product_Category_2'].fillna(Final_dataset['Product_Category_2'].mode()[0])
```
